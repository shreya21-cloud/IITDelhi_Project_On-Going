```python

!pip install -q transformers accelerate bitsandbytes pillow torch torchvision

import torch
from transformers import pipeline
from PIL import Image
from google.colab import files

print("GPU available:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
else:
    print("GPU not available. LLaVA may not run properly on CPU.")


print("Loading LLaVA model...")

pipe = pipeline(
    task="image-text-to-text",
    model="llava-hf/llava-1.5-7b-hf",
    device_map="auto",
    torch_dtype=torch.float16
)

print("LLaVA model loaded successfully!")

print("Upload an image:")
uploaded = files.upload()

image_path = list(uploaded.keys())[0]
image = Image.open(image_path).convert("RGB")

print("Image uploaded successfully!")
display(image)


print("\nVisual Question Answering Started")
print("Ask questions about the uploaded image.")
print("Type 'exit' to stop.\n")

while True:

    question = input("Question: ")

    if question.lower() == "exit":
        print("VQA stopped.")
        break

    prompt = f"USER: <image>\n{question}\nASSISTANT:"

    result = pipe(
        images=image,
        text=prompt,
        max_new_tokens=100
    )

    answer = result[0]["generated_text"]

    print("\nAnswer:")
    print(answer)
    print("-" * 50)
```
